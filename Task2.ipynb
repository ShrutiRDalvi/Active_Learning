Question 2


```python
import pandas as pd
from sklearn.datasets import load_iris

# Load the Iris dataset
iris = load_iris()# tried to laod using a different form compared to task 1
iris_df = pd.DataFrame(data=iris.data, columns=iris.feature_names)

# Display the original DataFrame
print("Original Iris DataFrame:")
print(iris_df.head())

# Specify the column and threshold for filtering
column_to_filter = 'sepal length (cm)'
threshold_value = 5.0

# Filter rows where 'sepal length (cm)' is greater than 5.0
filtered_iris_df = iris_df[iris_df[column_to_filter] > threshold_value]

# Display the filtered DataFrame
print(f"\nRows where '{column_to_filter}' > {threshold_value}:")
print(filtered_iris_df)
```

```python
import matplotlib.pyplot as plt
import seaborn as sns
from sklearn.datasets import load_iris

# Load the Iris dataset
iris = load_iris()
iris_df = sns.load_dataset("iris")

# Extracting sepal length values
sepal_length = iris_df['sepal_length']

# Plotting the line graph
plt.plot(sepal_length, label='Sepal Length', linestyle='-', marker='o')

# Customize the plot
plt.title('Line Graph of Sepal Length in Iris Dataset')
plt.xlabel('Index')
plt.ylabel('Sepal Length (cm)')
plt.legend()

# Display the plot
plt.show()
```

Question 4

```python
import pandas as pd
import numpy as np

# Create a DataFrame with random values including NaN
data = {
    'A': np.random.rand(15),
    'B': np.random.rand(15),
    'C': [np.nan, 2, np.nan, 4, 5, np.nan, 7, 8, 9, 1,2,3,4,5,6],
}

df = pd.DataFrame(data)

# Display the original DataFrame
print("Original DataFrame:")
print(df)

# Replace NaN values with the mean of the column
df = df.fillna(df.mean())

# Display the DataFrame after replacing NaN values
print("\nDataFrame after replacing NaN values with mean:")
print(df)

# Sort the DataFrame based on two columns (sorting by 'A' in ascending order, then by 'B' in descending order)
df_sorted = df.sort_values(by=['A', 'B'], ascending=[True, False])

# Display the sorted DataFrame
print("\nDataFrame sorted based on columns 'A' and 'B':")
print(df_sorted)
```

Question 5


```python
import numpy as np

# Define two matrices
matrix1 = np.array([[1, 2], [3, 4]])
matrix2 = np.array([[5, 6], [6, 9]])

# Calculate the dot product of the matrices
dot_product_result = np.dot(matrix1, matrix2)

# Find the determinant of the resulting matrix
determinant_result = np.linalg.det(dot_product_result)

print("\nDot Product Result:")
print(dot_product_result)

print("\nDeterminant of the Resulting Matrix:")
print(determinant_result)
```

```python
import pandas as pd
import seaborn as sns

# Load the Iris dataset
iris = sns.load_dataset("iris")

# Display the first few rows of the original DataFrame
print("Original DataFrame:")
print(iris.head())

# Group the DataFrame by the 'species' column and calculate the sum of 'sepal length (cm)' for each group
grouped_df = iris.groupby('species')['sepal_length'].sum().reset_index()

# Display the result
print("\nSum of 'sepal length (cm)' for each species:")
print(grouped_df)
```

```python
import matplotlib.pyplot as plt
import seaborn as sns

# Load the Iris dataset
iris = sns.load_dataset("iris")

# Extract the 'sepal length (cm)' column for the histogram
sepal_length_data = iris['sepal_length']

# Create a histogram
plt.hist(sepal_length_data, bins=20, color='skyblue', edgecolor='black')

# Customize the plot
plt.title('Histogram of Sepal Length in Iris Dataset')
plt.xlabel('Sepal Length (cm)')
plt.ylabel('Frequency')
plt.grid(True, linestyle='--', alpha=0.5)

# Display the plot
plt.show()
```

Question 9


```python
import pandas as pd

# Create the first DataFrame
data1 = {
    'Team': ['CSK', 'MI', 'RCB', 'KKR', 'SRH'],
    'City': ['Chennai', 'Mumbai', 'Bangalore', 'Kolkata', 'Hyderabad'],
    'Captain': ['MS Dhoni', 'Rohit Sharma', 'Virat Kohli', 'Eoin Morgan', 'Kane Williamson'],
    'HomeGround': ['M. A. Chidambaram Stadium', 'Wankhede Stadium', 'M. Chinnaswamy Stadium', 'Eden Gardens', 'Rajiv Gandhi International Stadium']
}

df1 = pd.DataFrame(data1)

# Create the second DataFrame
data2 = {
    'Team': ['CSK', 'MI', 'RCB', 'KKR', 'SRH'],
    'Coach': ['Stephen Fleming', 'Mahela Jayawardene', 'Simon Katich', 'Brendon McCullum', 'Trevor Bayliss'],
    'Owner': ['Chennai Super Kings Cricket Ltd', 'Reliance Industries', 'United Spirits', 'Knight Riders Sports Pvt. Ltd', 'Sun TV Network']
}

df2 = pd.DataFrame(data2)

#merge on 'Team' column
merged_df = pd.merge(df1, df2, on='Team')

print("DataFrame 1:")
print(df1)

print("\nDataFrame 2:")
print(df2)

print("\nMerged DataFrame:")
print(merged_df)
```

```python
import matplotlib.pyplot as plt
import seaborn as sns

# Load the Iris dataset
iris = sns.load_dataset("iris")

# Extract numerical variables for the scatter plot
sepal_length = iris['sepal_length']
sepal_width = iris['sepal_width']

# Create a scatter plot
plt.scatter(sepal_length, sepal_width, color='blue', marker='o', label='Sepal Data')

# Customize the plot
plt.title('Scatter Plot of Sepal Length vs Sepal Width')
plt.xlabel('Sepal Length (cm)')
plt.ylabel('Sepal Width (cm)')
plt.legend()
plt.grid(True, linestyle='--', alpha=0.7)

# Display the plot
plt.show()
```
